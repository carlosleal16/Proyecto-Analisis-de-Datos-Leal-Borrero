### 2. Crear el Dataset
Ejecuta la siguiente celda para crear los DataFrames que usaremos en los ejercicios.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo
sns.set_theme(style='whitegrid')
np.random.seed(42)

# Dataset 1: Ventas mensuales
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
ventas_2022 = [45, 52, 48, 65, 70, 85, 90, 88, 75, 80, 95, 110]
ventas_2023 = [50, 58, 55, 72, 78, 92, 98, 95, 82, 88, 105, 120]

df_ventas = pd.DataFrame({
    'Mes': meses,
    'Ventas_2022': ventas_2022,
    'Ventas_2023': ventas_2023
})

# Dataset 2: Datos de empleados
n_empleados = 150
df_empleados = pd.DataFrame({
    'Edad': np.random.normal(35, 8, n_empleados).astype(int).clip(22, 65),
    'Salario': np.random.normal(50000, 15000, n_empleados).clip(25000, 120000),
    'Años_Experiencia': np.random.normal(8, 4, n_empleados).clip(0, 30),
    'Departamento': np.random.choice(['Ventas', 'IT', 'Marketing', 'RRHH'], n_empleados),
    'Satisfaccion': np.random.randint(1, 11, n_empleados)
})

# Dataset 3: Productos
productos = ['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Auriculares']
unidades_vendidas = [120, 450, 380, 210, 290]

df_productos = pd.DataFrame({
    'Producto': productos,
    'Unidades': unidades_vendidas
})

print('Datasets creados exitosamente')
print(f'\ndf_ventas: {df_ventas.shape}')
print(f'df_empleados: {df_empleados.shape}')
print(f'df_productos: {df_productos.shape}')
```

### 3. Gráfico de Líneas Simple
Crea un gráfico de líneas que muestre las ventas de 2022 (`Ventas_2022`) a lo largo de los meses. Usa `plt.plot()` y añade título y etiquetas a los ejes.

```python
# Tu código aquí
plt.figure(figsize=(10, 6))
plt.plot(df_ventas['Mes'], df_ventas['Ventas_2022'])
plt.title('Ventas de 2022')
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
plt.savefig('mi_grafico3.png', dpi=400)
plt.show()



```

```python
df_ventas['Ventas_2022'].describe().T
```

De acuerdo a la gráfica anterior de ventas 2022 por meses, se evidencia lo siguiente:
- El promedio de ventas del 2022 fue de 77.5.
- Los meses de mayor auge son (julio, octubre, noviembre y diciembre) con un promedio de 92.5.
- Los meses de menores ventas fueron: Enero, febrero y marzo esto posiblemente debido a comienzo del año y temas escolares y postvacaciones.

### 15. Heatmap de Correlación
Crea un heatmap que muestre la matriz de correlación de las variables numéricas de `df_empleados`. Usa `sns.heatmap()` con `annot=True`.

```python
# Tu código aquí
plt.figure(figsize=(10,6))
correlacion_matrix = df_empleados[['Edad', 'Salario','Años_Experiencia','Satisfaccion']].corr(method='spearman')
sns.heatmap(correlacion_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

- La matriz de correlación muestra las relaciones entre varias variables: Edad, Salario, Años de Experiencia y Satisfacción:

- Edad vs. Salario: La correlación es -0.032, lo que indica una relación muy débil y negativa. Es decir, no existe una correlación significativa entre la edad y el salario.

- Edad vs. Años de Experiencia: La correlación es -0.029, lo que también indica una correlación negativa y muy débil, sugiriendo que no hay una relación directa entre la edad y los años de experiencia.

- Edad vs. Satisfacción: La correlación es 0.084, una correlación positiva débil, lo que sugiere que existe una pequeña relación positiva entre la edad y la satisfacción.

- Salario vs. Años de Experiencia: La correlación es 0.003, lo que indica que no existe ninguna correlación significativa entre el salario y los años de experiencia.

- Salario vs. Satisfacción: La correlación es -0.025, lo que también sugiere una relación muy débil y negativa entre el salario y la satisfacción.

### 17. Subplots (Múltiples Gráficos)
Crea una figura con 2 subplots (1 fila, 2 columnas):
- Subplot 1: Histograma de edades
- Subplot 2: Histograma de salarios

Usa `plt.subplots()`.

```python
# Tu código aquí
fig, axes = plt.subplots(2, 2, figsize=(15, 10))  # 2 filas, 2 columnas

# Histograma 1 (Edad)
axes[0, 0].hist(df_empleados['Edad'], bins=15)
axes[0, 0].set_title('Distribución de Edades')
axes[0, 0].set_xlabel('Edad')
axes[0, 0].set_ylabel('Frecuencia')

# Histograma 2 (Salario)
axes[0, 1].hist(df_empleados['Salario'], bins=15, color='orange')
axes[0, 1].set_title('Distribución de Salarios')
axes[0, 1].set_xlabel('Salario')
axes[0, 1].set_ylabel('Frecuencia')

# Histograma 3 (Edad)
axes[1, 0].hist(df_empleados['Edad'], bins=15)
axes[1, 0].set_title('Distribución de Edades')
axes[1, 0].set_xlabel('Edad')
axes[1, 0].set_ylabel('Frecuencia')

# Histograma 4 (Salario)
axes[1, 1].hist(df_empleados['Salario'], bins=15, color='blue')
axes[1, 1].set_title('Distribución de Salarios')
axes[1, 1].set_xlabel('Salario')
axes[1, 1].set_ylabel('Frecuencia')

plt.tight_layout()  # Ajusta los espacios entre las gráficas
plt.show()
```

- La mayoría de los empleados se concentran entre los 30 y 35 años, con una frecuencia más baja en edades mayores a 40. Esto sugiere que hay una mayor cantidad de empleados jóvenes en la organización.
- La mayoría de los salarios están en el rango de 40,000 a 60,000, con una caída significativa para salarios más altos. Esto indica que los salarios más comunes están en un rango medio, con pocos empleados ganando salarios elevados.
